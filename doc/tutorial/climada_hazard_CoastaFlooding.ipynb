```python
from climada_petals.hazard import CoastalFlood
```

Global costal flood maps for various RCPs, years and return periods are pulled from the acqueduct project : http://wri-projects.s3.amazonaws.com/AqueductFloodTool/download/v2/index.html

```python
rcps=['historical', 'rcp4p5']
years=['hist', '2080']
return_periods=['5', '1000']

cf = CoastalFlood().from_aqueduct_tif(
                                       rcps=rcps,
                                       years=years,
                                       return_periods=return_periods,
                                       boundaries = (88.5, 20, 92.5, 27)
                                       )
```

```python
cf.plot_intensity(event=0)
```
